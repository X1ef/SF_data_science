## 3. Pandas Series

### 3.1 Что такое Series?

Структура/объект Series представляет из себя объект, похожий на одномерный массив (питоновский список, например), но отличительной его чертой является наличие ассоциированных меток, т.н. индексов, вдоль каждого элемента из списка. Такая особенность превращает его в ассоциативный массив или словарь в Python.

### 3.2 Создание Series

```python
my_series = pd.Series(
    data=[5, 6, 7, 8, 9, 10],
    index=['a', 'b', 'c', 'd', 'e', 'f'],
    name='simple_series',
    dtype='float16'
)#создание из двух списков
my_series
```

```python
pd.Series()
```

### 3.3 Получение данных из Series

```python
my_series[['b', 'c', 'd']]
```

```python
print('Значения Series: ', list(my_series.values))
print('Индексы Series: ', list(my_series.index))
```

### 3.4 Фильтрация данных в Series

## 4. Pandas DataFrame

### 4.1. Что такое DataFrame?

DataFrame является двумерной структурой и представляется в виде таблицы, в которой есть строки и столбцы. Столбцами в DataFrame выступают объекты Series, а строки формируются из их элементов. Также в DataFrame есть метки (индексы), которые соответствуют каждой строке таблицы.

### 4.2. Создание DataFrame

```python
df = pd.DataFrame(data={
    'country': ['Kazakhstan', 'Russia', 'Belarus', 'Ukraine'],
    'population': [17.04, 143.5, 9.5, 45.5],
    'square': [2724902, 17125191, 207600, 603628]
},
#index=['a', 'b', 'c', 'd']
)#создание датафрейма из словаря
df
```

```python
df.index = ['KZ', 'RU', 'BY', 'UA']#зададим ассоциативные метки(индексы)
df
```

### 4.3. Доступ к данным в DataFrame

### 4.5. Создание новых столбцов в DataFrame

```python
df['population'] / df['square'] * 1000000
```

```python
# Создаем новую колонку 
df['density'] = df['population'] / df['square'] * 1000000
df
```

```python
#альтернативный вариант
density = []
for p, s in zip(df['population'], df['square']):
  density.append(p/s * 1000000)
df['density'] = density 
df
```

#### 5.2.1 Анализ числовых признаков

Сначала разберемся с **числовыми** характеристиками. 

Под числовыми признаками обычно подразумевают признаки, которые отражают количественную меру и могут принимать значения из неограниченного диапазона. Числовые признаки могут быть:
* дискретными, например, количество комнат в здании, пациентов, количество человек
* непрерывными: масса, цена, площадь

Такие признаки чаще всего кодируются целыми числами и числами, с плавающей точкой.

```python
amazon_df.head()
```

Для исследования числовых признаков можно использовать агрегирующие методы: mean



(), std(), max(), min(), count()

Найдем статистические характеристики для цен на товары и построим гистограмму:

```python
print('Средняя цена товаров (руб): ', amazon_df['price'].mean())
print('Медианная цена товаров (руб): ', amazon_df['price'].median())
print('Минимальная цена товаров (руб): ', amazon_df['price'].min())
print('Максимальная цена товаров (руб): ', amazon_df['price'].max())
print('Стандартное отклонение цены на товары (руб): ', amazon_df['price'].std())
amazon_df['price'].hist();
```

Найдем средний рейтинг товаров бренда Calvin Klein

```python
amazon_df[(amazon_df['brand_name'] == 'Calvin-Klein')]['rating'].mean()
```

#### 5.2.2. Анализ категориальных признаков

Под категориальными признаками обычно подразумевают столбцы в таблице, которые обозначают принадлежность объекта к какому-то классу/категории. Категориальные признаки могут быть:
* номинальными, например, пол, национальность, район
* порядковыми: уровень образования, уровень комфорта,стадия заболевания.

Такие признаки имеют ограниченный набор значений, они чаще всего представлены в виде текстового описания и кодируются в Pandas типом данных object.

#### *5.2.3 Методы группировки данных

Метод **groupby**


изображение_2021-03-20_163932.png

- Разделение данных по группап по определенному критерию

- Применение к каждой группе определенной функции

- Сложение результатов в единую структуру

Сравним бренды по средним ценам на их товары:



```python
#Получение информации по конкретной категории с помощью группировки
amazon_df.groupby('brand_name')['price'].mean().sort_values(ascending=False)
```

А теперь по ценам, рейтингам и числу отзывов одновременно:

```python
amazon_df.groupby(
    'brand_name'
)[[ 'rating', 'price', 'review_count']].mean().sort_values(by=['rating', 'price', 'review_count'], ascending=[False, True, False])
```

Найдем на какие бренды предлагают наибольшие скидки:

```python
amazon_df['sale'] = amazon_df['mrp'] - amazon_df['price']
amazon_df.groupby('brand_name')['sale'].mean().sort_values(ascending=False)
```
